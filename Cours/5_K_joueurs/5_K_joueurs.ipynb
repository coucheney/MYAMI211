### Modèl général

---

* ensemble fini de joueurs $\{1, \dots, K\}$
* chaque joueur $k$ a un ensemble fini d'actions $\mathcal{S}_k$
* chaque joueur a une fonction de gain (ou utilité) $U_k: \mathcal{S} \rightarrow \mathbb{R}$ qu'ils cherchent à maximiser, avec $\mathcal{S} = \mathcal{S}_1\times \mathcal{S}_2\times \dots \times \mathcal{S}_K$

* On note $\Delta_k$ l'ensemble des stratégies mixtes du joueur $k$.
* Etant donné le **profil de stratégies** $(p_1, p_2, \dots, p_k)$ avec $p_k \in \Delta_k$ pour tout $k \in \{1, \dots, K \}$, on étend la fonction de gain à un tel profil en considérant le gain moyen donné par
$$U_k(p_1, p_2, \dots, p_K) = \sum_{s \in \mathcal{S}}\mathbb{P}[S = s]U_k(s) $$
et 
$$ \mathbb{P}[S = (s_1, s_2, \dots, s_k)] = p_{1,s_1}p_{2,s_2}\dots p_{K,s_k}$$

### Terminologie

---

**Profil de stratégies**: donnée d'une stratégie par joueur. L'ensemble des profils de stratégie est noté $\Delta = \Delta_1\times \Delta_2\times \dots \times \Delta_K$


**Profil de stratégie adverse du joueur k**: ce sont les profils de stratégies qui ne contiennent pas la stratégie du joueur $k$. On note $\Delta_{-k}$ l'ensemble des stratégies adverses du joueur $k$, et $p_{-k}$ une telle stratégie.

Un profil de stratégie peut alors s'écrire $(p_k, p_{-k})$.

### Réponse optimale et équilibre de Nash

---


Etant donné un profil de stratégie adverse $p_{-k}$ pour le joueur $k$, une réponse optimale est une stratégie qui maximise la fonction $p_k \rightarrow U_k(p_k, p_{-k})$.

On note $\text{Opt}_k(p)$ l'ensemble des réponses optimales du joueur $k$ face au profil de stratégies adverses $p_{-k}$. Il s'agit d'une fonction multivaluée, ou **correspondance**.



Un **équilibre de Nash** (ou EN) est un profil de stratégies où aucun joueur n'a intérêt à changer de stratégie unilatéralement.

Autrement dit, $p\in \Delta$ est un EN si et seulement si 
 $$U_k(p_k, p_{-k}) \geq U_k(p'_k, p_{-k}), \forall k \in \{1, \dots, K \}, \forall p'_k \in \Delta_k$$
ou
$$p_k \in \text{Opt}_k(p), \forall k \in \{1, \dots, K \}$$
ou vectoriellement
$$p \in \text{Opt}(p) $$
donc si et seulement si $p$ est un **point fixe** de la correspondance $\text{Opt}$.

**Question**: Montrer que dans un jeu à deux joueurs et à somme nulle, les équilibres de Nash et les équilibres minimax coincident.


### Théorème de Nash

---

Dans un jeu sous forme normale fini, il existe toujours un EN.

La preuve repose sur le résultat suivant (admis).

**Théorème de Brouwer**: une fonction continue $f: C \rightarrow C$ où $C$ est un ensemble convexe et compact admet un point fixe.



<div class="alert-info">
Preuve du théorème de Nash:
</div>
   Consider la fonction continue $f: \Delta \rightarrow \Delta$ telle que pour tout joueur $k$ et action $i$ de ce joueur:
   $$f_{k,i}(p) = \frac{p_{k,i} + d_{k,i}(p)^+}{1 + \sum_j d_{k,j}(p)^+} $$
   où
   $$d_{k,i}(p) = U_k(i, p_{-k}) - U_k(p) $$
   et
   $$ x^+ = \max\{0, x\} $$
   
   Ses points fixes coïncident avec les équilibres de Nash du jeu.
   

### Support d'une stratégie

---

Le support d'une stratégie $p \in \Delta_k$ est l'ensemble des actions $i \in \mathcal{S}_k$ qui ont une probabilité non nulle d'être choisie, *i.e.* telles que $p_{i} > 0$. 

**Exemples**:
* $p = (0.3,0,0.1,0,0.4,0.2)$ a pour support les actions $\{1,3,5,6\}$,
* une stratégie pure est une stratégie dont le support est réduit à un seul élément.

**Propriété**: une stratégie $p$ du joueur $k$ est une réponse optimale à la stratégie adverse $p_{-k}$ si et seulement si toutes les actions qui sont dans son support sont des réponses optimales à $p_{-k}$.

**Exemple**: si un joueur fait face au vecteur de gains moyens $(4,0,4,2,-1,4)$, alors toute réponse optimale a pour support uniquement des actions de l'ensemble $\{1,3,6 \}$.

**Question**: démontrer cette propriété.

### Jeu à 2 joueurs: tester s'il existe un EN avec une paire de supports donnée
---

Soit $T_1 \subseteq S_1$ et $T_2 \subseteq S_2$ des supports des joueurs 1 et 2 respectivement. Alors tout  $(p_1,p_2)$ de l'ensemble défini par les **contraintes linéaires** suivantes est un EN (peut être vide):

* pour tout $s\in T_1$ et tout $s' \in S_1$,
    $$U_1(s, p_2)\geq U_1(s', p_2)$$ 

* pour tout $s\in T_2$ et tout $s' \in S_2$,
    $$U_2(p_1, s) \geq U_1(p_1, s')$$
    
* $p_{1, s} = 0$ si $s \notin T_1$

* $p_{2, s} = 0$ si $s \notin T_2$

* $p_1 \in \Delta_1$

* $p_2 \in \Delta_2$


### Jeu à 2 joueurs: calcul de l'ensemble des ENs par énumération des supports
---


Pour chaque support de chaque joueur:


* calculer les stratégies des adversaires qui égalisent les gains associés aux actions dans le support
* si elles existent, vérifier si ces gains sont bien optimaux par rapport aux actions en dehors du support
* si c'est le cas, retourner ces stratégies

**Exemple**:

$$
\left(
\begin{array}{c|c}
3,3& 3, 2\\
\hline
2, 2&5, 6\\
\hline
0,3 & 6,1  \\
\end{array}
\right)
$$

Le nombre de supports pour chaque joueur ayant $n$ actions est $2^n-1$.

L'algorithme de Lemke-Howson permet de trouver un EN par méthode de pivotage du type simplexe, mais sa complexité reste exponentielle.

### Stratégie strictement dominée

---

Une stratégie pure $s_k$ du joueur $k$ est strictement dominée s'il existe une stratégie $p_k$ qui donne un gain strictement supérieur quel que soit le profil de stratégie adverse, *i.e.* $\forall p_{-k} \in \Delta_{-k}, U_k(s_k, p_{-k}) < U_k(p_k, p_{-k})$.



**Propriété**: une stratégie strictement dominée n'est dans le support d'aucun EN.

Les stratégies strictement dominées peuvent être supprimées itérativement (**algorithme  d'élimination des stratégies strictement dominées**).

**Exemple**:

$$
\left(
\begin{array}{c|c|c}
3,3& 2, 2 & 3,0\\
\hline
2, 3&6, 5 & 1,4\\
\hline
1,4 & 3,4 & 1.5,5  \\
\end{array}
\right)
$$


**Questions**:
* démontrer la propriété
* donner un exemple qui montre qu'elle n'est pas vrai si on remplace la  domination stricte par une domination faible (gain inférieur ou égal).

### Quelques résultats de complexité

---


* calcul d'un EN d'un jeu à 2 joueurs: PPAD-complet, aussi difficile que le problème du calcul d'un point fixe de Brouwer.
* les problèmes de décision suivants sont NP-complets:
    * existe-t'il au moins 2 ENs?
    * existe-t'il un EN qui garantit une valeur $v$ au joueur $i$?
* résoudre un jeu à somme nulle à 3 joueurs est aussi difficile que de résoudre un jeu à 2 joueurs à somme quelconque.
* un jeu à 2 joueurs symétrique admet un EN symétrique.
* calculer un EN dans un jeu à 2 joueurs symétrique est aussi difficile qu'un jeu non symétrique.
